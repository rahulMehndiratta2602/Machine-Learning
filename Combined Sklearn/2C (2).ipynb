**Importing Necessary Libraries and Dataset**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn import neighbors,metrics
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Perceptron
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


df = pd.read_csv('/content/drive/MyDrive/ML Assignment2/dataset_comb.csv')
```

```python
def grouping(Class):
    if Class == 'jasmine':
        return 0
    elif Class == 'Gonen':
        return 1
df['Class'] = df['Class'].apply(grouping)

df = df.drop(['id'], axis = 1)
df = df.sample(frac = 1)
df = df.sample(frac = 1)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

**SVM**

```python
testaccuracy_svm = []
trainaccuracy_svm = []


kf = KFold(n_splits = 7)
for train_index,test_index in kf.split(X):
     X_train, X_test = X.iloc[train_index], X.iloc[test_index]
     y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     sc = StandardScaler()
     X_train = sc.fit_transform(X_train)
     X_test = sc.transform(X_test)
     model = svm.SVC()
     model.fit(X_train,y_train)
     prediction_test = model.predict(X_test)
     prediction_train = model.predict(X_train)
     acc_test = accuracy_score(y_test,prediction_test)
     acc_train = accuracy_score(y_train,prediction_train)
     testaccuracy_svm.append(acc_test)
     trainaccuracy_svm.append(acc_train)

print("Mean accuracy - svm for test set = ", np.mean(testaccuracy_svm))
print("Mean accuracy - svm for train set = ", np.mean(trainaccuracy_svm))
```

**Naives Bayes Classifier**

```python
testaccuracy_nb = []
trainaccuracy_nb = []

kf = KFold(n_splits = 7)
for train_index,test_index in kf.split(X):
     X_train, X_test = X.iloc[train_index], X.iloc[test_index]
     y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     sc = StandardScaler()
     X_train = sc.fit_transform(X_train)
     X_test = sc.transform(X_test)
     cl_Gaussian = GaussianNB()
     model = cl_Gaussian.fit(X_train, y_train)
     prediction_test = model.predict(X_test)
     acc_test = accuracy_score(y_test, prediction_test)
     prediction_train = model.predict(X_train)
     acc_train = accuracy_score(y_train,prediction_train)
     testaccuracy_nb.append(acc_test)
     trainaccuracy_nb.append(acc_train)

print("Mean accuracy - nb for test set = ", np.mean(testaccuracy_nb))
print("Mean accuracy -nb for train set = ", np.mean(trainaccuracy_nb))
```

**Logistic Regression**

```python
testaccuracy_lr = []
trainaccuracy_lr = []

kf = KFold(n_splits = 7)
for train_index,test_index in kf.split(X):
     X_train, X_test = X.iloc[train_index], X.iloc[test_index]
     y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     sc = StandardScaler()
     X_train = sc.fit_transform(X_train)
     X_test = sc.transform(X_test)
     cl = LogisticRegression(random_state=0, max_iter=1000)
     model = cl.fit(X_train, y_train)
     prediction_test = model.predict(X_test)
     acc_test = accuracy_score(y_test, prediction_test)
     prediction_train = model.predict(X_train)
     acc_train = accuracy_score(y_train,prediction_train)
     testaccuracy_lr.append(acc_test)
     trainaccuracy_lr.append(acc_train)

print("Mean accuracy -lr for test set = ", np.mean(testaccuracy_lr))
print("Mean accuracy -lr for train set = ", np.mean(trainaccuracy_lr))
```

**Fisher Linear Discriminant**

```python
testaccuracy_fda = []
trainaccuracy_fda = []

kf = KFold(n_splits = 7)
for train_index,test_index in kf.split(X):
     X_train, X_test = X.iloc[train_index], X.iloc[test_index]
     y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     sc = StandardScaler()
     X_train = sc.fit_transform(X_train)
     X_test = sc.transform(X_test)
     cl = LinearDiscriminantAnalysis()
     model = cl.fit(X_train, y_train)
     prediction_test = model.predict(X_test)
     acc_test = accuracy_score(y_test, prediction_test)
     prediction_train = model.predict(X_train)
     acc_train = accuracy_score(y_train,prediction_train)
     testaccuracy_fda.append(acc_test)
     trainaccuracy_fda.append(acc_train)

print("Mean accuracy -fda for test set = ", np.mean(testaccuracy_fda))
print("Mean accuracy -fda for train set = ", np.mean(trainaccuracy_fda))
```

**Linear** **Perceptron**

```python
testaccuracy_lp = []
trainaccuracy_lp = []

kf = KFold(n_splits = 7)
for train_index,test_index in kf.split(X):
     X_train, X_test = X.iloc[train_index], X.iloc[test_index]
     y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     sc = StandardScaler()
     X_train = sc.fit_transform(X_train)
     X_test = sc.transform(X_test)
     cl = Perceptron(tol = 1e-3, random_state=0)
     model = cl.fit(X_train, y_train)
     prediction_test = model.predict(X_test)
     acc_test = accuracy_score(y_test, prediction_test)
     prediction_train = model.predict(X_train)
     acc_train = accuracy_score(y_train,prediction_train)
     testaccuracy_lp.append(acc_test)
     trainaccuracy_lp.append(acc_train)

print("Mean accuracy -lp for test set = ", np.mean(testaccuracy_lp))
print("Mean accuracy -lp for train set = ", np.mean(trainaccuracy_lp))
```

**Artificial Neural Network**

```python
testaccuracy_ann = []
trainaccuracy_ann = []

kf = KFold(n_splits = 7)
for train_index,test_index in kf.split(X):
     X_train, X_test = X.iloc[train_index], X.iloc[test_index]
     y_train, y_test = y.iloc[train_index], y.iloc[test_index]
     sc = StandardScaler()
     X_train = sc.fit_transform(X_train)
     X_test = sc.transform(X_test)
     cl = MLPClassifier(hidden_layer_sizes = (10, 10, 10), max_iter = 1000)
     model = cl.fit(X_train, y_train)
     prediction_test = model.predict(X_test)
     acc_test = accuracy_score(y_test, prediction_test)
     prediction_train = model.predict(X_train)
     acc_train = accuracy_score(y_train,prediction_train)
     testaccuracy_ann.append(acc_test)
     trainaccuracy_ann.append(acc_train)

print("Mean accuracy for -ann test set = ", np.mean(testaccuracy_ann))
print("Mean accuracy for -ann train set = ", np.mean(trainaccuracy_ann))
```

```python

fig = plt.figure(figsize = (20, 10))
plt.boxplot([testaccuracy_lr, testaccuracy_lp, testaccuracy_svm, testaccuracy_nb, testaccuracy_fda, testaccuracy_ann])

plt.text(0.75, 0.96, "Logistic Regression")
plt.text(0.80, np.mean(testaccuracy_lr) , "Mean: " + str(round(np.mean(testaccuracy_lr), 5)))

plt.text(1.75, 0.96, "Linear Perceptron")
plt.text(1.82, np.mean(testaccuracy_lp) , "Mean: " + str(round(np.mean(testaccuracy_lp), 5)))

plt.text(2.70, 0.96, "Support Vector Machine")
plt.text(2.82, np.mean(testaccuracy_svm) , "Mean: " + str(round(np.mean(testaccuracy_svm), 5)))

plt.text(3.85, 0.96, "Naive Bayes")
plt.text(3.80, np.mean(testaccuracy_nb) , "Mean: " + str(round(np.mean(testaccuracy_nb), 5)))

plt.text(4.68, 0.96, "Fisher Linear Discriminant")
plt.text(4.80, np.mean(testaccuracy_fda) , "Mean: " + str(round(np.mean(testaccuracy_fda), 5)))

plt.text(5.68, 0.96, "Artificial Neural Networks")
plt.text(5.80, np.mean(testaccuracy_ann), "Mean: " + str(round(np.mean(testaccuracy_ann), 5)))
plt.show()

```
